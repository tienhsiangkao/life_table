```python
import pandas as pd
import matplotlib.pyplot as plt

usa = pd.read_csv (r'usa.csv')
roc = pd.read_csv (r'roc.csv')
usa['2019'] = pd.to_numeric(usa['2019'],errors='coerce')
roc['2019'] = pd.to_numeric(roc['2019'],errors='coerce')
usa['2019.1'] = pd.to_numeric(usa['2019.1'],errors='coerce')
roc['2019.1'] = pd.to_numeric(roc['2019.1'],errors='coerce')
usa['2019.2'] = pd.to_numeric(usa['2019.2'],errors='coerce')
roc['2019.2'] = pd.to_numeric(roc['2019.2'],errors='coerce')
usa['2015'] = pd.to_numeric(usa['2015'],errors='coerce')
roc['2015'] = pd.to_numeric(roc['2015'],errors='coerce')
usa['2015.1'] = pd.to_numeric(usa['2015.1'],errors='coerce')
roc['2015.1'] = pd.to_numeric(roc['2015.1'],errors='coerce')
usa['2015.2'] = pd.to_numeric(usa['2015.2'],errors='coerce')
roc['2015.2'] = pd.to_numeric(roc['2015.2'],errors='coerce')
usa['Unnamed: 1'] = usa['Unnamed: 1'].map(lambda x: x.rstrip('years'))
roc['Unnamed: 1'] = roc['Unnamed: 1'].map(lambda x: x.rstrip('years'))
```

```python
fig0 = plt.figure(dpi=400)
plt.plot(usa['Unnamed: 1'][-18:], usa['2019'][-18:], label='USA', color = 'b')
plt.plot(roc['Unnamed: 1'][-18:], roc['2019'][-18:], label='China', color = 'r')
plt.legend()
plt.ylabel('life expectancy');
plt.xlabel('age')
plt.grid()
plt.xticks(fontsize=5)
plt.title('USA vs China')
```

```python
fig1 = plt.figure(dpi=400)
plt.plot(usa['Unnamed: 1'][-18:], usa['2019'][-18:], label='Both Sexes')
plt.plot(usa['Unnamed: 1'][-18:], usa['2019.1'][-18:], label='Male')
plt.plot(usa['Unnamed: 1'][-18:], usa['2019.2'][-18:], label='Female')

plt.legend()
plt.ylabel('life expectancy');
plt.xlabel('age')
plt.grid()
plt.xticks(fontsize=5)
plt.title('USA: Sexes')
```

```python
fig2 = plt.figure(dpi=400)
plt.plot(roc['Unnamed: 1'][-18:], roc['2019'][-18:], label='Both Sexes')
plt.plot(roc['Unnamed: 1'][-18:], roc['2019.1'][-18:], label='Male')
plt.plot(roc['Unnamed: 1'][-18:], roc['2019.2'][-18:], label='Female')

plt.legend()
plt.ylabel('life expectancy');
plt.xlabel('age')
plt.grid()
plt.xticks(fontsize=5)
plt.title('China: Sexes')
```

```python
fig3 = plt.figure(dpi=400)
plt.plot(usa['Unnamed: 1'][-18:], usa['2019.2'][-18:]-usa['2019.1'][-18:], label='USA', color='b')

plt.plot(roc['Unnamed: 1'][-18:], roc['2019.2'][-18:]-roc['2019.1'][-18:], label='China', color='r')

plt.legend()
plt.ylabel('life expectancy difference');
plt.xlabel('age')
plt.grid()
plt.xticks(fontsize=5)
plt.title('USA vs China: Sexes')
```

```python
fig3 = plt.figure(dpi=400)
plt.plot(usa['Unnamed: 1'][-18:], usa['2019'][-18:]-usa['2015'][-18:], label='USA', color='b')

plt.plot(roc['Unnamed: 1'][-18:], roc['2019'][-18:]-roc['2015'][-18:], label='China', color='r')

plt.legend()
plt.ylabel('life expectancy change');
plt.xlabel('age')
plt.grid()
plt.xticks(fontsize=5)
plt.title('USA vs China: 2015-2019')
```
